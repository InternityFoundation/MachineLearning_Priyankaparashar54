# Support Vector Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

### Printing matrix of features X and dependent variable y

```python
print(X)
```

```python
print(y)
```

#### Reshape the dependent variable y

```python
#y = y.reshape(len(y),1)
y= y.reshape(-1, 1)
```

Reshaping the dependent variable y into an array because we have to use standard scalar class that will perform standardization that means feature scaling expects one unique format in it's input. When we apply fit transform method which is a 2-D array, if you input here 1-D vector like that we have it will give us error simply because standard scalar class expects it's input to be 2-D aaray. So we have reshape it.

```python
print(y)
```

## Encoding categorical data

```python
#we are dummy encoding as the machine learning algorithms will be
#confused with the values like New York > Florida > California

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Splitting the dataset into train set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:].astype(float))
X_test[:,3:]=sc.transform(X_test[:,3:].astype(float))
y_train[:,:]=sc.fit_transform(y_train[:,:].astype(float))
y_test[:,:]=sc.fit_transform(y_test[:,:].astype(float))
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Training the Support Vector Regression model on the Training set

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf',gamma='auto')
regressor.fit(X_train,np.ravel(y_train,order='C'))
```

 ## Accuracy of train model

```python
print(f'r_square value: {regressor.score(X_train, y_train)}')
```

## Predicting the test set results

```python
y_pred = regressor.predict(X_test)
```

## Accuracy of test model

```python
print(f'r_square value: {regressor.score(X_test, y_test)}')
```
